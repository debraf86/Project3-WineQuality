```python
import pandas as pd
```

```python
# read the csv files - 1 per year
dfRedWine = pd.read_csv("./Resources/winequality-red.csv")
dfRedWine.head()
```

```python
# iterate through each row and normalize between 0 and 1. 
df=dfRedWine

for i in range(len(dfRedWine)) : 
    df.loc[i, "fixed acidity"] = round((dfRedWine.loc[i, "fixed acidity"] - dfRedWine["fixed acidity"].min())/ (dfRedWine["fixed acidity"].max()-dfRedWine["fixed acidity"].min()),2)
    df.loc[i, "volatile acidity"] = round((dfRedWine.loc[i, "volatile acidity"] - dfRedWine["volatile acidity"].min())/ (dfRedWine["volatile acidity"].max()-dfRedWine["volatile acidity"].min()),2)
    df.loc[i, "sulphates"] = round((dfRedWine.loc[i, "sulphates"] - dfRedWine["sulphates"].min())/ (dfRedWine["sulphates"].max()-dfRedWine["sulphates"].min()),2)
    df.loc[i, "density"] = round((dfRedWine.loc[i, "density"] - dfRedWine["density"].min())/ (dfRedWine["density"].max()-dfRedWine["density"].min()),2)
    df.loc[i, "residual sugar"] = round((dfRedWine.loc[i, "residual sugar"] - dfRedWine["residual sugar"].min())/ (dfRedWine["residual sugar"].max()-dfRedWine["residual sugar"].min()),2)
    df.loc[i, "free sulfur dioxide"] = round((dfRedWine.loc[i, "free sulfur dioxide"] - dfRedWine["free sulfur dioxide"].min())/ (dfRedWine["free sulfur dioxide"].max()-dfRedWine["free sulfur dioxide"].min()),2)
    df.loc[i, "total sulfur dioxide"] = round((dfRedWine.loc[i, "total sulfur dioxide"] - dfRedWine["total sulfur dioxide"].min())/ (dfRedWine["total sulfur dioxide"].max()-dfRedWine["total sulfur dioxide"].min()),2)
    df.loc[i, "pH"] = round((dfRedWine.loc[i, "pH"] - dfRedWine["pH"].min())/ (dfRedWine["pH"].max()-dfRedWine["pH"].min()),2)
    df.loc[i, "alcohol"] = round((dfRedWine.loc[i, "alcohol"] - dfRedWine["alcohol"].min())/ (dfRedWine["alcohol"].max()-dfRedWine["alcohol"].min()),2)
    df.loc[i, "quality"] = round((dfRedWine.loc[i, "quality"] - dfRedWine["quality"].min())/ (dfRedWine["quality"].max()-dfRedWine["quality"].min()),2)
```

```python
df.head()
```

```python
# write the normalized data to a new csv file.
df.to_csv(r'./normalizedWineData.csv')
```
